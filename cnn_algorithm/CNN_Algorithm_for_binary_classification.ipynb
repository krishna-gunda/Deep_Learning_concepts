```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
```

```python
path='/content/sample_parrot-image.jpg'
image=cv2.imread(path)
kernal_filter=np.array([[0,1,0],[1,-4,1],[0,1,0]])
result=cv2.filter2D(image,-1,kernel=kernal_filter)

cv2_imshow(image)
cv2_imshow(result)
```
