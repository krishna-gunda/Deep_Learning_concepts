###  Objective
In this notebook, we explore different image color spaces using OpenCV.

We will:
- Load an image
- Convert it into different color formats:
  - RGB
  - HSV
  - Grayscale
- Display and understand the differences

---

###  Why Color Spaces?
Different color spaces are used for different purposes:
- **RGB** → Displaying images correctly
- **HSV** → Color detection and segmentation
- **Grayscale** → Simplifies image (used in edge detection, ML models)

```python
import cv2
import numpy as np
```

```python
from google.colab.patches import cv2_imshow
```

###  Load the Image
We read the image using OpenCV.
By default, OpenCV loads images in **BGR format**.

###  Convert BGR → RGB

OpenCV uses BGR format, but most systems use RGB.
So we convert it for correct color representation.

###  Convert BGR → HSV

HSV stands for:
- Hue → Color type
- Saturation → Color intensity
- Value → Brightness

HSV is mainly used in:
- Color detection
- Object tracking

###  Convert BGR → Grayscale

Grayscale removes color information and keeps only intensity.

Used in:
- Edge detection
- Face detection
- Machine learning models

```python
image=cv2.imread('/content/sample_parrot-image.jpg')
rgb_image=cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
hsv_image=cv2.cvtColor(image,cv2.COLOR_BGR2HSV)
gray_scale_image=cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
cv2_imshow(image)# showimg original image
cv2_imshow(rgb_image) # showing RGB color space image
cv2_imshow(hsv_image) # showing hsv color space image
cv2_imshow(gray_scale_image) # converting into black image
```

###  Summary

| Conversion | Purpose |
|----------|--------|
| BGR → RGB | Correct display |
| BGR → HSV | Color-based processing |
| BGR → Grayscale | Simplifies image |

###  Color Detection using HSV (Red Color)

In this section, we detect **red color** from the image using HSV color space.

---

###  Steps Explained:

1. **Convert Image to HSV**
   - OpenCV reads image in BGR format
   - HSV is better for color detection

2. **Define Color Range**
   - `lower_bound` and `higher_bound` define the range of red color in HSV

3. **Create Mask**
   - `cv2.inRange()` checks each pixel:
     - If pixel is within range → **white (255)**
     - Else → **black (0)**
   - Output is a **binary mask**

4. **Apply Mask**
   - `cv2.bitwise_and()` keeps only red parts
   - Non-red areas become black

---

###  Output:

- **Original Image** → Input image  
- **HSV Image** → Used for processing  
- **Mask** → White = red detected  
- **Result** → Only red color visible  

---

###  Key Idea:
👉 Detect color → Create mask → Apply mask → Get desired color

---

###  Note:
Red color has two ranges in HSV, so for better detection we may need to combine two masks.

```python
# now iam trying to detect specific color in the hsv image

image=cv2.imread('/content/sample_parrot-image.jpg') # reading the original parrot image

hsv_image=cv2.cvtColor(image,cv2.COLOR_BGR2HSV) # changing original image to hsv image

lower_bound=np.array([0, 120, 70]) # red lower bound values
higher_bound=np.array([10, 255, 255]) # red higher bound values

red_mask=cv2.inRange(hsv_image,lower_bound,higher_bound)  # marking only the redcolor in the image

result=cv2.bitwise_and(image,image,mask=red_mask)

cv2_imshow(image)
cv2_imshow(hsv_image)
cv2_imshow(red_mask)
cv2_imshow(result)
```
