Image Blurring Techniques in OpenCV (Beginner Friendly Guide)
Introduction

Image blurring is a technique used in image processing to reduce noise and details in an image. It smooths the image by modifying pixel values using their neighboring pixels.

In simple words:
Blurring = making an image less sharp and more smooth

Blurring is widely used in computer vision for:

Noise removal
Preprocessing before edge detection
Improving image quality
Types of Blurring Techniques

There are mainly three important blurring techniques:

Gaussian Blur
Median Blur
Bilateral Filter
1. Gaussian Blur
What it does

Gaussian blur smooths the image using a weighted average of neighboring pixels.

Easy Understanding

Imagine spreading colors with a soft brush
Everything becomes smooth, but edges also become blurry

Code Example
gaussian = cv2.GaussianBlur(image, (21, 21), 0)
Key Points
Produces natural smoothing
Reduces noise effectively
Widely used in preprocessing
2. Median Blur
What it does

Median blur replaces each pixel with the median value of its neighboring pixels instead of averaging.

Easy Understanding

Imagine removing random black and white dots from an image
The image becomes clean while keeping edges clear

Code Example
median = cv2.medianBlur(image, 11)
Key Points
Best for salt-and-pepper noise
Preserves edges better than Gaussian
Non-linear filtering method
3. Bilateral Filter
What it does

Bilateral filter smooths the image while preserving edges by considering both pixel distance and intensity difference.

Easy Understanding

Imagine smoothing only similar colors while keeping boundaries sharp
Edges remain clear while flat areas become smooth

Code Example
bilateral = cv2.bilateralFilter(image, 9, 75, 75)
Key Points
Preserves edges while smoothing
Works based on color and distance
More advanced but slower

```python
from cv2.gapi import gaussianBlur
image=cv2.imread('/content/sample_parrot-image.jpg')
cvt_image=cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
gaussian=cv2.GaussianBlur(cvt_image,(201,201),3)

cv2_imshow(image) # showing the original image
cv2_imshow(gaussian) # showing the blur image
```
