````python
#converting color image to Grayscale image (black image)
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
````

````python
# reading the image
image = cv2.imread('/content/thumb-1920-1368954.jpeg')
#  Loads the image from the given file path
#  Stores it as a NumPy array (matrix of pixel values)
#  Image is read in BGR (Blue, Green, Red) format by default

result = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
#  Converts the color image into grayscale (black & white)
#  cv2.cvtColor() is used for color space conversion
#  COLOR_BGR2GRAY tells OpenCV to convert from BGR → Gray
#  Output will have only 1 channel (intensity)

cv2_imshow(image)
#  Displays the original color image
#  Used in Google Colab instead of cv2.imshow()

cv2_imshow(result)
#  Displays the grayscale image
#  Helps to visually compare original vs processed image
````

'Method 2: Using the cv2.imread() function with flag=zero'

````python
image = cv2.imread('/content/thumb-1920-1368954.jpeg', 0)
#  Reads the image from the given file path
#  The flag `0` tells OpenCV to load the image in grayscale mode
#  Output image will have only 1 channel (no color, only brightness values)

cv2_imshow(image)
#  Displays the grayscale image
#  Used in Google Colab instead of cv2.imshow()
#  The image will appear in black and white (shades of gray)
````
